```python
def parsedict(feature):
    fv=feature.tolist()
    nfv=[]
    for i,item in enumerate(fv):
        d={}
        if type(item) == str:
            tmp= item.split('||')
            if len(tmp)==1:
                tmp=item.split('|')
            for t in tmp:
                tmp2= t.split('::')
                if len(tmp2)==1:
                    tmp2=item.split(':')
                d[tmp2[0]]=tmp2[1]
            nfv.append(d)
        else:
            nfv.append(item)
    processed_feature=pd.Series(nfv)
    return processed_feature
        
```

```python
def process_by_labels(pdata):
    charcterstics=list(item.replace('||','|').split('|') for item in pdata['incident_characteristics'].dropna(axis=0).tolist())
    labels=[]
    for incident in charcterstics:
        for item in incident:
            if item not in labels:
                labels.append(item)
    incident_type=np.zeros((len(charcterstics),len(labels)))
    for j,incident in enumerate(charcterstics):
        for i,inctyp in enumerate(labels):
            if inctyp in incident:
                incident_type[j][i]=1
    chtyp=pd.DataFrame(incident_type,columns=labels, dtype=bool)
    res=chtyp.apply(pd.value_counts)
    return res,labels
```

```python
def retrieve_by_label(nature):
    inds=[]
    copydata=processed_data.copy()
    for ind,i in enumerate(copydata['incident_characteristics'].dropna(axis=0)):
        if nature in i:
            inds.append(ind)
    remlst=list(set(copydata.index.values)-set(copydata.index.values[inds]))
    copydata=copydata.drop(remlst)
    return copydata
```

```python
##Grouping to Miscellaneous
def form_group_miscellaneous(resl,t,labls):
    nlabls=labls[:]
    N=sum(resl.iloc[1])
    ctr=0
    misc_sum=0
    for item in resl.iloc[1]:
        if float(item)/N < t:
            misc_sum+=item
            resl=resl.drop(labls[ctr],axis=1)
            nlabls.remove(labls[ctr])
        ctr=ctr+1
    nlabls.append('Miscellaneous')
    resl['Miscellaneous']=pd.Series([0,misc_sum])
    return resl,nlabls
```

```python
## Grouping all incident types that occur less than 1% of the times into miscellaneous
All_res,labels=process_by_labels(processed_data)
All_res,labels=form_group_miscellaneous(All_res,0.01,labels)
```

## Participant Analysis

```python
def participant_analysis(pdata):
    suspects_status=[]
    victims_status=[]
    suspects_age=[]
    victims_age=[]
    ps=[]
    pt=[]
    pstatus=pdata['participant_status'].dropna(axis=0)
    pagegroup=pdata['participant_age_group'].dropna(axis=0)
    ptype=pdata['participant_type'].dropna(axis=0)
    inds=ptype.index.values
    inds2=pstatus.index.values
    inds3=pagegroup.index.values
    inds_status=list(set(inds).intersection(set(inds2)))
    for i in inds_status:
            itemst=pstatus.loc[i]
            itmept=ptype.loc[i]
            for i in itmept.keys():
                if itmept[i]=='Subject-Suspect':
                    if i in itemst.keys():
                        suspects_status.append(itemst[i])
                    else:
                        suspects_status.append('Unknown')
                elif itmept[i]=='Victim':
                    if i in itemst.keys():
                        victims_status.append(itemst[i])
                    else:
                        victims_status.append('Unknown')   
    
    inds_age=list(set(inds).intersection(set(inds3)))
    for i in inds_age:
            itemag=pagegroup.loc[i]
            itmept=ptype.loc[i]
            for i in itmept.keys():
                if itmept[i]=='Subject-Suspect':
                    if i in itemag.keys():
                        suspects_age.append(itemag[i])
                    else:
                        suspects_age.append('Unknown')
                elif itmept[i]=='Victim':
                    if i in itemag.keys():
                        victims_age.append(itemag[i])
                    else:
                        victims_age.append('Unknown')   
    return suspects_status,victims_status,suspects_age,victims_age   
```

```python
def plot_suspect_victim_bar(cnt_sus,cnt_vic,typ,cat):
    sup = go.Bar(
                x=cnt_sus.index,
                y=list(cnt_sus),
                name='Subject-Suspects'
        )
    vic = go.Bar(
                x=cnt_vic.index,
                y=list(cnt_vic),
                name='Victims'
        )
    dat=[sup,vic]
    layout = go.Layout(
        barmode='group',
        title='Suspect - Victim '+ typ + ' Chart- ' + cat 
    )

    fig = go.Figure(data=dat, layout=layout)
    display(py.iplot(fig, filename='suspects-victims-'+ typ + '-' + cat))
```
